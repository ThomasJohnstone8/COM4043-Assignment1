# Task 1: Cloud/Hosting Solution Review

*A small local bakery wants to set up an online store for click-and-collect orders. They are unsure whether to pick a traditional shared web hosting package (like from GoDaddy) or a cloud-based platform (like AWS or Azure) to run a simple website.*


## Costs


Firstly, we must compare the costs differences of either solution. Shared hosting is the first option which comes with a simple monthly fee which can range from £5-£30 per month depending on what commodities you choose such as back-ups, storage, etc. 

Then you have cloud hosting typically run by AWS (Amazon Web Service) or Microsoft Azure which operate as a pay as you go method. meaning that the business will only pay for what they are using. This leads me on to my next point scalability. As this payment model can get expensive the bigger the business it is important to know what is right for the company 

 
## Scalability

Secondly, the scalability of a shared host can differ, if it is a sudden spike in user traffic, the website can be overloaded and quickly run into problems however, if it is a gradual increase, the payment plan can be quickly changed to scale to the needs of the business. 

On the other hand, cloud hosting is great with scalability, services such as AWS allow you to scale to the needs of your business depending on the demand of the user. The trick with cloud hosting is allocating more servers to deal with the increase of customers.  


## Ease of Setup

The next point to make is with ease of setup. Firstly, shared hosting is much easier to set up, having a comfortable UI to follow on a website selecting the commodities you need for the website is easy. Additionally, having a simple customer support line where the user can get first hand support if they come across any issues. 

Cloud hosting is a more complex setup as it is a more technical solution compared to a simple UI on a website. A team of IT professionals would be needed to set up and configure a cloud hosting solution which would be more time consuming. 

 
## Maintenance
 

Lastly, maintenance and upgrades are the last point to consider. Shared hosting maintenance is often covered by the provider, this is a large advantage as it is one less thing to need to worry about. 

Cloud hosting on the other hand has more responsibilities when it come to maintenance. One example is that software updates, security patches and more will need to be handled by the company itself meaning that the business would need to invest into a management team or have someone tech savvy. 


# Final Outcome

After weighting both the positives and negatives I believe the best option for the business is the shared hosting. One reason for this is the shared hosting solution has less responsibility and that means it's one less thing the business needs to deal with. Another reason is scalability of a shared host is that the plan can be upgraded quickly and simply making it suitable to smaller businesses. However, one downside to this is that if a sudden growth happens for example the holidays, then the business cannot upgrade for that short window. Lastly the maintenance of a shared hosting plan is covered by the provider saving the business the hassle of figuring out how to update security, software changes and more.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts

data = np.random.normal(0, 1, 1000)

mean = np.mean(data)
median = np.median(data)
std = np.std(data)
skew = sts.skew(data)
kurt = sts.kurtosis(data)

x = np.linspace(min(data), max(data), 1000)


print(f"Mean: {mean:.2f}")
print(f"Median: {median:.2f}")
print(f"Standard deviation: {std:.2f}")
print(f"Skewness: {skew:.2f}")
print(f"Kurtosis: {kurt:.2f}")

plt.hist(data, bins=30, density=True, alpha=0.6, color='g')
plt.plot(x, sts.norm.pdf(x, mean, std), 'r', linewidth=2)
plt.title('Task 3')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend
plt.grid()
plt.show()
```
